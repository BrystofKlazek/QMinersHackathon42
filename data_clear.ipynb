```python
import seaborn as sns
import matplotlib.pyplot as plt

corr = df.corr()

plt.figure(figsize=(10, 8))
plt.imshow(corr, interpolation='nearest')
plt.colorbar()
plt.xticks(range(len(corr)), corr.columns, rotation=90)
plt.yticks(range(len(corr)), corr.columns)
plt.title("Correlation Matrix Heatmap", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
dependent = []  # start with a list
independent = []

for i in range(len(corr)):
    for j in range(i+1):  # iterate over lower triangle
        if i != j:
            if corr.iloc[i, j] < -0.5 or corr.iloc[i, j] > 0.5:
                dependent.append((corr.columns[i], corr.columns[j]))  # append tuple
            else:
                independent.append((corr.columns[i], corr.columns[j]))

print(dependent)
```

```python
from scipy.stats import wilcoxon # dependent
from scipy.stats import mannwhitneyu # independent

throw_away = []

for el in dependent:
    stat, p = wilcoxon(df[el[0]], df[el[1]])
    if p > 0.05:
        throw_away.append(el)
print(throw_away)

for el in independent:
    stat, p = mannwhitneyu(df[el[0]], df[el[1]])
    if p > 0.05:
        throw_away.append(el)
print(throw_away)

#zat√≠m asi rad≈°i nemazat nic
```
